```python
import pandas as pd
```

```python
df_cities = pd.read_csv('cities500.txt', sep='\t', header=0)
```

```python
df_cities
```

```python
df_new = pd.DataFrame()
df_new['name'] = df_cities['asciiname']
df_new['population'] = df_cities['population']
df_new['timezone'] = df_cities['timezone']
continent_list = []
country_list = []
for i in range(0,len(df_new['timezone'])):
    T = df_new['timezone'][i].split('/')
    continent_list.append(T[0])
    country_list.append(T[1])
df_new['continent'] = continent_list
df_new['country'] = country_list
df_new = df_new.drop(['timezone'], axis=1)
df_new.to_csv('dataset_continent_country.csv',index=False)
```
